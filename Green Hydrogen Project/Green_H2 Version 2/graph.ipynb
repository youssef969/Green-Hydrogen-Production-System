```python
data = [70,20,10,50,60]
labels = ["Python","Java","HTML","C#","CCS"] 
colors = ["red","Green","blue","yellow","orange"] # الوان
explode = [0,0,0.1,0,0] # دي علشان ابين اي المهم او القيمة الي عاوزها تبقى بره شوية 

plt.pie(data,labels=labels,colors=colors,
        wedgeprops={"edgecolor":"Black"},
        explode=explode, 
        shadow=True,  # دي علشان تعمل ظل
        startangle=0, # دي علشان تقول اول سلايس هتبدا من زواية كام 
        autopct="%1.0f%%", # دي علشان تطلع النسب وعلى حسب الرقم الي بعد الفاصلة هتطلع كام عدد عشري
        rotatelabels=True  # ودي علشان العناوين تبقى داير ما يدور يعني 
        
        )
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib.figure import Figure
import random 
from tkinter import *

root = Tk()
root.geometry("500x400+100+100")

x_values = []
y_values = []

for _ in range(1000):
    x_values.append(random.randint(0,100))
    y_values.append(random.randint(0,100))
    
    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.scatter(x_values,y_values,color="red")
    plt.pause(0.001)

plt.show(block=True)
root.mainloop()
```

```python
import tkinter as tk
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import random
from time import strftime
import datetime
root = tk.Tk()
root.title("Live Data Plot with Shifting in Tkinter")


fig = Figure(figsize=(5, 4), dpi=100)
plot = fig.add_subplot(111)
line, = plot.plot([], [],label="y-data",color="black",linewidth=3)
line2, = plot.plot([],[],color="r",label="v-data",marker="o",linestyle="--")# Empty line for initial plot

x_data = []
y_data = []
v_data = []

canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
canvas.get_tk_widget().place(x=10,y=10,width=1000,height=800)
plot.grid()
#plt.style.use("_classic_test_patch")

def update_plot():
    x_data.append(len(x_data)+1)
    y_data.append(random.randint(1, 10))
    v_data.append(random.randint(1,20))
    plot.set_title("Green")
    
    line.set_data(x_data, y_data)
    line2.set_data(x_data, v_data)
    plot.set_ylim(0,40)
    #plot.set_xlim(0,100)
    plot.set_xlabel("Hello")
    plot.set_ylabel("Hello")
    plot.set
    plot.legend()
    plot.relim()
    plot.autoscale_view()
    fig.canvas.draw_idle()
    root.after(1000, update_plot)  


update_plot()


root.mainloop()
```
